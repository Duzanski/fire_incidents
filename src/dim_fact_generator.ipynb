```python
import pandas as pd

pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

# Loading Raw Table

```python
raw_table = pd.read_parquet("/Users/reinaldoduzanskijunior/Documents/teste/fire_incidents/extract_data/raw_database.parquet", columns=["incident_number", "incident_date", "neighborhood_district", "battalion", "primary_situation", "station_area", "fire_fatalities", "civilian_fatalities", "supervisor_district", "estimated_property_loss"])
raw_table.head(10)
```

# Start building Dim_Battalion

```python
dim_battalion = raw_table[['battalion', 'station_area']].drop_duplicates().reset_index(drop=True)
dim_battalion.head(10)
```

```python
dim_battalion['battalion_id'] = dim_battalion.index + 1
dim_battalion.head(10)
```

# Saving Dim_Battalion

```python
dim_battalion.to_parquet("/Users/reinaldoduzanskijunior/Documents/teste/fire_incidents/data_models/dimension_tables/dim_battalion_fire_incidents.parquet")
```

# Start building Dim_District

```python
dim_district = raw_table[['neighborhood_district', 'supervisor_district']].drop_duplicates().reset_index(drop=True)
dim_district.head(10)
```

```python
dim_district['district_id'] = dim_district.index + 1
dim_district.head(10)
```

# Saving Dim_District

```python
dim_district.to_parquet("/Users/reinaldoduzanskijunior/Documents/teste/fire_incidents/data_models/dimension_tables/dim_district_fire_incidents.parquet")
```

# Start building Dim_Time


```python
raw_table["incident_date"] = pd.to_datetime(raw_table["incident_date"])
raw_table['year'] = raw_table["incident_date"].dt.year
raw_table['month'] = raw_table["incident_date"].dt.month
raw_table['day'] = raw_table["incident_date"].dt.day
raw_table['day_of_week'] = raw_table["incident_date"].dt.dayofweek
raw_table.head(10)
```

```python
dim_time = raw_table[['incident_date', 'year', 'month', 'day', 'day_of_week']].drop_duplicates().reset_index(drop=True)
dim_time.head(10)
```

```python
dim_time['date_id'] = dim_time.index + 1
dim_time.head(10)
```

# Saving Dim_Date

```python
dim_time.to_parquet("/Users/reinaldoduzanskijunior/Documents/teste/fire_incidents/data_models/dimension_tables/dim_time_fire_incidents.parquet")
```

# Start building Fact_Table

```python
raw_table
```

```python
fact_fire_incidents = raw_table.merge(dim_battalion[['battalion', 'battalion_id']], on='battalion')
fact_fire_incidents = fact_fire_incidents.merge(dim_time[['incident_date', 'date_id']], on='incident_date')
fact_fire_incidents = fact_fire_incidents.merge(dim_district[['neighborhood_district', 'district_id']], on='neighborhood_district')
```

```python
fact_fire_incidents.head(10)
```

```python
fact_fire_incidents = fact_fire_incidents[['incident_number', 'date_id', 'district_id', 
                                           'battalion_id', 'primary_situation', 'station_area', 'fire_fatalities', 'civilian_fatalities', 'estimated_property_loss']]
fact_fire_incidents.head(10)
```

```python
fact_fire_incidents['estimated_property_loss'] = fact_fire_incidents['estimated_property_loss'].fillna(0)
fact_fire_incidents.head(10)
```

```python
fact_fire_incidents = fact_fire_incidents.astype({
    'incident_number': 'int64',      
    'date_id': 'int64',               
    'district_id': 'int64',            
    'battalion_id': 'int64',           
    'primary_situation': 'string',    
    'station_area': 'string',          
    'fire_fatalities': 'int64',       
    'civilian_fatalities': 'int64',    
    'estimated_property_loss': 'float' 
})
fact_fire_incidents.head(10)
```

```python
fact_fire_incidents.to_parquet("/Users/reinaldoduzanskijunior/Documents/teste/fire_incidents/data_models/fact_table/fact_fire_incidents.parquet")
```
